## Example
- Let's consider [https://en.wikipedia.org/wiki/Wikipedia:Fundraising_statistics](https://en.wikipedia.org/wiki/Wikipedia:Fundraising_statistics)
- **pandas** ```.read_html(.)``` Read HTML tables into a list of DataFrame objects ([docs](https://pandas.pydata.org/docs/reference/api/pandas.read_html.html)).

````python
import pandas as pd
````

````python
url = "https://en.wikipedia.org/wiki/Wikipedia:Fundraising_statistics"

data = pd.read_html(url)
````

````python
fund = data[0]
````

````python
fund.dtypes
````

````python
fund['Expenses'] = fund['Expenses'].str[2:]
````

````python
fund['Expenses'] = fund['Expenses'].str.replace(',','')
````

````python
fund['Expenses'] = pd.to_numeric(fund['Expenses'])
````

````python
fund.head()
````

## Data Wrangling
- Data wrangling (data munging): transforming and mapping data from one "raw" data form into another format
- With the intent of making it more appropriate and valuable for a variety of downstream purposes such as analytics

### Check the data types
- Remember ```.dtypes```

````python
fund
````

````python
fund['Asset rise'] = fund['Asset rise'].str[2:]
fund['Asset rise'] = fund['Asset rise'].str.replace(',','')
fund['Asset rise'] = pd.to_numeric(fund['Asset rise'])
````

````python
fund.head()
````

````python
fund['Total assets'] = fund['Total assets'].str[2:]
fund['Total assets'] = fund['Total assets'].str.replace(',','')
fund['Total assets'] = pd.to_numeric(fund['Total assets'])
````

````python
fund.head()
````

````python
fund['Year'] = fund['Year'].str.replace('-20','-')
````
